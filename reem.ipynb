```python
q1=df.price.quantile(0.25)
q3=df.price.quantile(0.75)
iqr= q3-q1
q1,q3,iqr

```

```python
lower_limit=q1-1.5*iqr
upper_limit=q3+1.5*iqr
lower_limit,upper_limit
```

```python
df[(df.price<lower_limit)|(df.price>upper_limit)]
```

```python
df_new=df[(df.price>lower_limit)&(df.price<upper_limit)]
df_new
```

```python
sns.boxplot(df_new['price']).set_title('Outlier1');
```

```python
plt.plot(df_new['number_of_reviews']);
```

```python
sns.boxplot(df_new['number_of_reviews']).set_title('Outlier');
```

```python
q11=df_new.number_of_reviews.quantile(0.25)
q33=df_new.number_of_reviews.quantile(0.75)
iqr2= q33-q11
q11,q33,iqr2
```

```python
lower_limit_n=q11-1.5*iqr2
upper_limit_n=q33+1.5*iqr2
lower_limit_n,upper_limit_n
```

```python
df_new[(df_new.number_of_reviews<lower_limit_n)|(df_new.number_of_reviews>upper_limit_n) ]
```

```python
df1=df_new[(df_new.number_of_reviews>lower_limit_n)&(df_new.number_of_reviews<upper_limit_n) ]
df1
```

```python
sns.boxplot(df1['number_of_reviews']).set_title('Outlier');
```

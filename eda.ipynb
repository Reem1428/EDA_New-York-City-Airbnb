# outlier

```python
df.price.describe()
```

```python
plt.plot(df['price']);
```

```python
sns.boxplot(df['price']).set_title('Outlier');
```

```python
q1=df.price.quantile(0.25)
q3=df.price.quantile(0.75)
iqr= q3-q1
q1,q3,iqr

```

```python
lower_limit=q1-1.5*iqr
upper_limit=q3+1.5*iqr
lower_limit,upper_limit
```

```python
df[(df.price<lower_limit)|(df.price>upper_limit)]
```

```python
df_new=df[(df.price>lower_limit)&(df.price<upper_limit)]
df_new
```

```python
sns.boxplot(df_new['price']).set_title('Outlier1');
```

```python
plt.plot(df_new['number_of_reviews']);
```

```python
sns.boxplot(df_new['number_of_reviews']).set_title('Outlier');
```

```python
q11=df_new.number_of_reviews.quantile(0.25)
q33=df_new.number_of_reviews.quantile(0.75)
iqr2= q33-q11
q11,q33,iqr2
```

```python
lower_limit_n=q11-1.5*iqr2
upper_limit_n=q33+1.5*iqr2
lower_limit_n,upper_limit_n
```

```python
df_new[(df_new.number_of_reviews<lower_limit_n)|(df_new.number_of_reviews>upper_limit_n) ]
```

```python
df1=df_new[(df_new.number_of_reviews>lower_limit_n)&(df_new.number_of_reviews<upper_limit_n) ]
df1
```

```python
sns.boxplot(df1['number_of_reviews']).set_title('Outlier');
```

# Does the price increase or decrease according to the neighborhood?

```python

y =df1.neighbourhood_group
x =df1.price

plt.figure(figsize=(8,4))
sns.set_theme(context='notebook', style='dark')

plt.xlabel('neighbourhood group',size = 15)
plt.ylabel('Price', size = 15)

plt.title("Price for each neighbourhood group ", size=20)

sns.barplot(y,x,color='blue',ci=None);
```

# Which type of room do visitors prefer?

```python
plt.title("type of room do visitors prefer ", size=20)
sns.barplot(df1['room_type'],df1['number_of_reviews'],color='blue',ci=None);
```

# Do neighborhood and prices affect the evaluation?

```python
fig, (ax1, ax2) = plt.subplots(1,2,figsize=(10,5))
fig.suptitle('number of reviews')
ax1.scatter(x='price',y='number_of_reviews',data=df1.sample(1000))
ax2.scatter('neighbourhood_group','number_of_reviews',data=df1.sample(1000));
ax1.set_title('price');
ax2.set_title('neighbourhood group');
#ax3.scatter('number_of_reviews','availability_365',data=df1.head(1000));
```
